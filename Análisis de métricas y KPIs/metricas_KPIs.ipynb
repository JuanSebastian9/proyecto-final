```python
# Suponiendo que ya cargaste todos los DataFrames
# Merge básico entre combine y player info
df_players = pd.merge(df_draft_combine_statsr_limpio, df_common_player_info, how='inner', left_on='player_id', right_on='person_id')

# Filtrar los jugadores recientes (últimos años de draft)
#df_players = df_players[df_players['season'] >= 2015]

# Graficar altura vs. rendimiento en salto vertical
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_players, x='height_w_shoes', y='max_vertical_leap', hue='position_x', palette='viridis')
plt.title('Altura vs. Salto Vertical por Posición (Draft Combine)')
plt.xlabel('Altura con zapatos (pulgadas)')
plt.ylabel('Salto vertical máximo (pulgadas)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Comparar agilidad vs. velocidad
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_players, x='lane_agility_time', y='three_quarter_sprint', hue='position_x', palette='plasma')
plt.title('Agilidad vs. Velocidad por Posición')
plt.xlabel('Tiempo agilidad (segundos)')
plt.ylabel('Sprint 3/4 cancha (segundos)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Distribución del peso por posición
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_players, x='position_x', y='weight_y', palette='Set2')
plt.title('Distribución de Peso por Posición')
plt.xlabel('Posición')
plt.ylabel('Peso (libras)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_other_stats['pts_paint_home'], kde=True, color='blue', label='Home', bins=20)
sns.histplot(df_other_stats['pts_paint_away'], kde=True, color='red', label='Away', bins=20)
plt.title('Distribución de puntos en la pintura - Local vs Visitante')
plt.xlabel('Puntos en la pintura')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df_other_stats[['pts_2nd_chance_home', 'pts_2nd_chance_away']])
plt.title('Boxplot - Puntos por segunda oportunidad')
plt.ylabel('Puntos')
plt.xticks([0,1], ['Local', 'Visitante'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_turnovers_home', y='pts_off_to_away', data=df_other_stats, color='darkgreen')
plt.title('Turnovers locales vs puntos del rival por turnovers')
plt.xlabel('Turnovers del equipo local')
plt.ylabel('Puntos del visitante generados por TO')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=[df_other_stats['team_rebounds_home'], df_other_stats['team_rebounds_away']],
               palette=["skyblue", "salmon"])
plt.title('Distribución de rebotes totales - Local vs Visitante')
plt.xticks([0, 1], ['Rebotes Locales', 'Rebotes Visitantes'])
plt.ylabel('Cantidad de rebotes')
plt.show()
```

```python
# Asegúrate de que los ID sean del mismo tipo
df_common_player_info['person_id'] = df_common_player_info['person_id'].astype(str)
df_draft_combine_statsr_limpio['player_id'] = df_draft_combine_statsr_limpio['player_id'].astype(str)

# Unir ambos DataFrames
df_players = pd.merge(df_draft_combine_statsr_limpio, df_common_player_info, left_on='player_id', right_on='person_id', how='inner')
```

```python
print(df_players.columns)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='position_x', y='height_w_shoes', data=df_players, palette='coolwarm')
plt.title('Altura con zapatos según posición')
plt.xlabel('Posición')
plt.ylabel('Altura (pulgadas)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='max_vertical_leap', y='lane_agility_time', hue='position_x', data=df_players, palette='viridis')
plt.title('Relación entre salto vertical y agilidad por posición')
plt.xlabel('Salto vertical máximo (pulgadas)')
plt.ylabel('Tiempo de agilidad en línea (segundos)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='position_x', y='bench_press', data=df_players, palette='Set2')
plt.title('Distribución del press de banca por posición')
plt.xlabel('Posición')
plt.ylabel('Repeticiones')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,10))
features = ['height_w_shoes', 'weight_y', 'wingspan', 'standing_vertical_leap', 'max_vertical_leap', 'lane_agility_time', 'three_quarter_sprint', 'bench_press']
sns.heatmap(df_players[features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación entre pruebas físicas')
plt.show()
```
